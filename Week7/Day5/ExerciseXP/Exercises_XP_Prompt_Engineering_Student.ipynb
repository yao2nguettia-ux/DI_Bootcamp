## 🌟 Exercise 1: Rewrite and Optimize a Vague Prompt

### Scenario:
You’re an AI Prompt Engineer at a productivity startup. Your manager asks you to use ChatGPT to create a LinkedIn post promoting their new focus app, FlowNest. They send you this vague prompt:

“Write something about productivity tips.”

### Tasks:

1.  **Identify 3 critical issues with this prompt.**
    (Hint: Think about brand voice, target users, and platform engagement.)

    **Issues:**
    1.  **Lack of specific audience/platform context:** The prompt doesn't specify who the LinkedIn post is for (e.g., busy professionals, students, remote workers) or the conventions of a LinkedIn post (e.g., professional tone, call to action).
    2.  **Missing brand voice/product integration:** It doesn't mention the product (FlowNest) or the company's brand voice. A generic response won't promote the app effectively.
    3.  **Vague objective/engagement:** The goal is to promote an app, but the prompt only asks for "something about productivity tips." It doesn't guide the AI to generate engaging content or include a call to action.


2.  **Rewrite the prompt using all three techniques:**
    *   Role prompting (e.g., “Act as a…”)
    *   Instruction-based prompting (e.g., “Give 3 actionable…”)
    *   Format/tone/length constraint (e.g., “Use bullet points” or “max 280 characters”)

    **Rewritten Prompt:**

    "Act as a social media marketing specialist for FlowNest, a new focus app for busy professionals. Draft a concise LinkedIn post (maximum 150 words) that shares 3 actionable productivity tips. Ensure the tone is professional, encouraging, and subtly integrates the benefits of using a focus app like FlowNest. Include a clear call to action to learn more about FlowNest."

3.  **Add a final line that makes the prompt reusable (e.g., “Use this format whenever we promote productivity tools on social media”).**

    **Reusable Prompt:**

    "Act as a social media marketing specialist for FlowNest, a new focus app for busy professionals. Draft a concise LinkedIn post (maximum 150 words) that shares 3 actionable productivity tips. Ensure the tone is professional, encouraging, and subtly integrates the benefits of using a focus app like FlowNest. Include a clear call to action to learn more about FlowNest. Use this format whenever we promote productivity tools or apps on LinkedIn."

## 🌟 Exercise 2: Multi-Part Prompt for Quiz Generation

### Scenario:
A 7th-grade science teacher uploads a short article on volcanic eruptions. You are tasked with writing a prompt for ChatGPT to create engaging educational content.

### Tasks:
1.  **Write a complete prompt that asks the model to:**
    *   Summarize the article in 2 simple bullet points
    *   Generate 3 age-appropriate multiple-choice questions, each with 1 correct and 2 distractors
    *   Include a friendly tone and simple vocabulary for 11–13-year-olds
    *   Output everything in a format suitable for Google Slides

2.  **Compare your prompt to this vague one:**
    “Make a quiz for kids about this article.”

3.  **Explain three key ways your version prevents irrelevant or low-quality output.**

### Solution for Exercise 2

**Complete Prompt:**

"Act as a friendly 7th-grade science teacher. I've provided an article about volcanic eruptions below. Please read it and then create engaging educational content for my 11-13 year old students.

Your output should follow these instructions:

1.  **Summary:** Provide a summary of the article in exactly two simple, easy-to-understand bullet points.
2.  **Multiple-Choice Questions:** Generate three multiple-choice questions about the article. Each question must have one correct answer and two plausible incorrect answers (distractors). Ensure the language and concepts are appropriate for 7th graders. Clearly indicate the correct answer for each question.
3.  **Tone & Vocabulary:** Maintain a friendly and encouraging tone throughout. Use simple vocabulary suitable for students aged 11-13.
4.  **Format for Google Slides:** Present the summary and questions in a clear, concise format that can be easily copied and pasted into Google Slides. Use clear headings for the summary and each question.

Here is the article:

[Insert Article Text Here]

"

**Comparison and Explanation:**

Comparing my detailed prompt to the vague prompt: “Make a quiz for kids about this article.”

My version prevents irrelevant or low-quality output in three key ways:

1.  **Clarity on Purpose and Audience (Role Prompting & Tone):** The vague prompt gives no context about *who* the quiz is for or *why* it's being made. My prompt explicitly states, "Act as a friendly 7th-grade science teacher" and specifies the content is for "11-13 year old students," guiding the AI to use appropriate language, complexity, and a friendly tone. This prevents the AI from generating content that is too advanced, too simple, or in an inappropriate style.

2.  **Specific Output Structure and Constraints (Instruction-based & Format/Length):** The vague prompt offers no guidance on the structure or quantity of the output. My prompt uses instruction-based prompting and format constraints by asking for "exactly two simple, easy-to-understand bullet points" for the summary and "three multiple-choice questions, each with one correct answer and two plausible incorrect answers (distractors)." It also specifies "Output everything in a format suitable for Google Slides." This prevents the AI from creating an unorganized, lengthy, or incomplete response, ensuring it delivers exactly what's needed for the educational context.

3.  **Defined Content Components (Instruction-based):** The vague prompt only asks for a "quiz." My prompt breaks down the required content into specific components: a summary *and* multiple-choice questions. It further defines what each question should contain (1 correct, 2 distractors). This specificity ensures all desired elements are present and correctly structured, rather than just a generic set of questions that might not be effective for teaching.

## 🌟 Exercise 4: Match Prompt to Purpose

### Scenario:
You’re building prompt templates for a customer support chatbot at an e-commerce company. Pick one of the following styles and use it to write a complete prompt for the chatbot.

*   **Exploratory:** Helps the model brainstorm options or generate ideas
*   **Structured:** Follows a fixed format (lists, sections, steps)
*   **Conversational:** Mimics a human, chatty tone
*   **Functional:** Executes a task with precision (e.g., summarizing or transforming input)

### Tasks:
1.  **Write your prompt using the selected style.**
2.  **Identify 2 key features that make your prompt a match for the style.**
3.  **Justify your choice: Why is this style better than the others for your specific case?**

### Solution for Exercise 4

**Selected Style:** Functional

**1. Prompt:**

"You are a customer support chatbot for an e-commerce company. Your task is to process customer inquiries regarding order status. When a customer provides an order number, extract the order number and check its status. If the order number is valid, return the current status (e.g., 'Processing', 'Shipped', 'Delivered') and the estimated delivery date if available. If the order number is invalid or not found, politely inform the customer and ask them to verify their order number. Do not generate additional conversational filler beyond what is strictly necessary to convey the order status or request clarification.

**Example Input:** 'Hi, I want to know about my order 12345.'
**Example Output:** 'Order 12345 is currently Shipped. Estimated delivery date: October 26, 2023.'

**Example Input:** 'What's the status of order 98765?'
**Example Output:** 'I apologize, but order number 98765 could not be found. Please double-check your order number and try again.'

**Customer Inquiry:** [Insert Customer Inquiry Here]"

**2. Key Features Matching the Style:**

*   **Precision and Task-Oriented:** The prompt clearly defines a specific task (processing order status inquiries) and outlines the exact steps and conditions for fulfilling it (extract order number, check status, return specific information or error). It leaves little room for ambiguity.
*   **Strict Output Control:** The prompt explicitly states what *should* be included in the output (status, delivery date, invalid message) and, importantly, what *should not* be included ("Do not generate additional conversational filler"). This ensures the output is direct and serves its precise function without unnecessary embellishments.

**3. Justification of Choice:**

The Functional style is better than the others for a customer support chatbot at an e-commerce company because:

*   **Efficiency:** Customers engaging with a support chatbot typically want quick, accurate answers to their problems. A functional prompt ensures the chatbot delivers precise information directly, minimizing conversational detours and getting straight to the point. This improves the customer experience by reducing frustration and wait times.
*   **Accuracy and Consistency:** For tasks like checking order status, accuracy is paramount. A functional prompt, with its strict instructions and constraints, reduces the likelihood of the chatbot hallucinating information, providing irrelevant details, or deviating from its core purpose. This ensures consistent and reliable service, which is crucial for maintaining customer trust.
*   **Scalability and Maintainability:** Functional prompts are easier to standardize and implement across various support scenarios. Their clear, step-by-step nature makes them simpler to test, debug, and update as business needs or data sources change. This is highly beneficial for managing a large-scale customer service operation.

## 🌟 Exercise 5: Prompt Refinement Challenge – Control the Style, Structure, and Length

### Scenario:
You work for a wearable tech company launching a new smartwatch called PulseOne Mini. The marketing team is preparing content for an email campaign and wants you to use ChatGPT to write a product blurb with these constraints:

*   Friendly tone
*   Bullet point format
*   Max 50 words total
*   Must mention battery life, fitness tracking, and Bluetooth compatibility

### Tasks:
1.  **Write a prompt that strictly enforces these constraints.**

2.  **Run the prompt and evaluate:**
    *   Did the model stick to the word count?
    *   Were all 3 product features mentioned?
    *   Is the tone friendly and engaging?

3.  **If not, revise your prompt with stronger constraint language (e.g., “Do not exceed 50 words. List exactly 3 bullet points.”)**

### Solution for Exercise 5

**1. Initial Prompt:**

"Write a product blurb for the new PulseOne Mini smartwatch for an email marketing campaign. The tone should be friendly. Present the information in bullet points. The total blurb should be no more than 50 words. Make sure to include mentions of battery life, fitness tracking, and Bluetooth compatibility."

**2. Evaluation (Example AI Output & Analysis):**

*Example AI Output (Hypothetical):*

> "Introducing the amazing PulseOne Mini! Get ready to experience unparalleled convenience. Enjoy incredible battery life that lasts for days, advanced fitness tracking to crush your goals, and seamless Bluetooth compatibility for all your devices. Stay connected and healthy with PulseOne Mini!"

*Analysis:*

*   **Did the model stick to the word count?** No. (e.g., 47 words, which is close but could be over in a real scenario, or it might generate more if the constraints aren't strict enough)
*   **Were all 3 product features mentioned?** Yes (battery life, fitness tracking, Bluetooth compatibility).
*   **Is the tone friendly and engaging?** Yes.

*Critique:* The initial prompt is good, but the word count constraint might not be strictly adhered to by all models without more forceful language.

**3. Revised Prompt with Stronger Constraint Language:**

"Generate a marketing blurb for the new PulseOne Mini smartwatch, targeting an email campaign.

**Constraints:**
*   **Tone:** Friendly and engaging.
*   **Format:** Strict bullet points only. Create exactly three distinct bullet points.
*   **Length:** The *entire* blurb, including all text and bullet points, MUST NOT EXCEED 50 words. Prioritize conciseness.
*   **Content:** Each of the following features MUST be mentioned once across the bullet points: extended battery life, comprehensive fitness tracking, and seamless Bluetooth compatibility.

Craft the blurb to be impactful within these limits."

## 🌟 Exercise 6: Hallucination Spotting and Mitigation

### Scenario:
You’re preparing a summary of a peer-reviewed article on climate change and marine biodiversity. ChatGPT inserts this claim:

“Over 50% of marine species are projected to go extinct by 2050.”

This claim does not appear in the original article.

### Tasks:
1.  **Write a precise prompt instructing the model to avoid hallucinations and use only the provided text.**

2.  **Rewrite the prompt using verification language, like:**
    *   “Only include claims explicitly stated in the article.”
    *   “If uncertain, state that the information is not available.”

3.  **List 2 mitigation strategies you used to reduce hallucination risk.**

4.  **Name 2 professional domains (e.g., legal, healthcare) where hallucinations are especially risky, and explain why.**

### Solution for Exercise 6

**1. Precise Prompt to Avoid Hallucinations:**

"Summarize the following peer-reviewed article on climate change and marine biodiversity. Your summary must STRICTLY adhere to the information contained within the provided text. Do not introduce any outside information, assumptions, or inferences. Every statement in your summary must be directly verifiable by a passage in the article.

[Insert Article Text Here]"

**2. Prompt with Verification Language:**

"Summarize the following peer-reviewed article on climate change and marine biodiversity. Only include claims explicitly stated in the article. If a piece of information is not directly verifiable or the article does not provide specific data for a statement, you MUST state that the information is not available in the provided text, or rephrase it to reflect the article's exact wording. Do not infer or extrapolate. Focus on the main findings and conclusions presented within the document.

[Insert Article Text Here]"

**3. Two Mitigation Strategies Used:**

1.  **Strict Source Adherence:** By using phrases like "STRICTLY adhere to the information contained within the provided text" and "Every statement in your summary must be directly verifiable by a passage in the article," I explicitly instruct the model to treat the provided text as the sole source of truth, severely limiting its ability to generate ungrounded claims.
2.  **Explicit Verification and Uncertainty Handling:** The prompt includes instructions for how to handle missing or uncertain information ("If uncertain, state that the information is not available"). This prevents the model from fabricating details when the source doesn't support them and encourages transparency about its knowledge limitations relative to the provided context.

**4. Two Professional Domains Where Hallucinations Are Especially Risky:**

1.  **Legal Domain:** Hallucinations in the legal domain are extremely risky because incorrect information (e.g., fabricated case precedents, misquoted laws, or non-existent statutes) can lead to severe consequences. Lawyers and judges rely on accurate legal research; an AI hallucinating a legal precedent could result in faulty advice, incorrect rulings, financial losses, or even wrongful convictions. The integrity of the justice system depends on factual accuracy.
2.  **Healthcare/Medical Domain:** In healthcare, AI-generated hallucinations could have life-threatening implications. Imagine an AI providing incorrect diagnostic information, recommending a non-existent drug, or fabricating side effects or dosages. This could lead to misdiagnosis, inappropriate treatments, adverse drug interactions, or harm to patients. The critical nature of patient safety demands absolute factual accuracy in medical information.
