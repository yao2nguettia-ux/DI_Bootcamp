## Step 1: Craft a Prompt That Controls Output Tone, Format, and Length

I will create a prompt that rewrites the given text in a friendly and clear tone, uses bullet points, paraphrases the original content, and stays under 75 words.

```python
input_text = """Employees must ensure that all remote access to internal systems is established via the approved secure VPN. Under no circumstances should unsecured connections or personal devices lacking endpoint protection be used to access proprietary data or sensitive communications."""

user_prompt = f"""Rewrite the following text for internal employee communication. The tone should be friendly and clear. Organize the information using bullet points. Paraphrase the original content, avoiding direct quotations. Keep the total output under 75 words.

Text to rewrite: {input_text}"""

print("Input Text:")
print(input_text)
print("\n--- User Prompt ---\n")
print(user_prompt)
```

### Setup Gemini API

To use the Gemini API, you'll need an API key. If you don't already have one, create a key in Google AI Studio. In Colab, add the key to the secrets manager under the "🔑" in the left panel. Give it the name `GOOGLE_API_KEY`. Then pass the key to the SDK:

```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

Now, initialize the Generative Model. The prompt for this challenge was tested with `gemini-pro`.

```python
# Initialize the Gemini API
gemini_model = genai.GenerativeModel('gemini-pro')
```

```python
response = gemini_model.generate_content(user_prompt)
print("\n--- Gemini Model Response ---\n")
print(response.text)
```
