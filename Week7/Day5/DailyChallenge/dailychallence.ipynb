## Step 1: Craft a Prompt That Controls Output Tone, Format, and Length

Here is the original input text:

**Input Text:**
"Employees must ensure that all remote access to internal systems is established via the approved secure VPN. Under no circumstances should unsecured connections or personal devices lacking endpoint protection be used to access proprietary data or sensitive communications."


Your task is to write a prompt that:
* Rewrites the text in a friendly and clear tone.
* Uses bullet points to organize information.
* Paraphrases the original, avoiding direct quotations.
* Stays under 75 words total.


Below, create your prompt and then use the `gemini_model.generate_content()` function to get the AI's output. Then print the output.
```

````python
input_text = "Employees must ensure that all remote access to internal systems is established via the approved secure VPN. Under no circumstances should unsecured connections or personal devices lacking endpoint protection be used to access proprietary data or sensitive communications."

# Craft the prompt
prompt = f"""Rewrite the following text in a friendly and clear tone.
Organize the information using bullet points.
Paraphrase the original text, avoiding direct quotations.
Ensure the total output stays under 75 words.

Text to rewrite: {input_text}
"""

# Generate content using the Gemini model
response = gemini_model.generate_content(prompt)

# Print the output
print(response.text)
````

## Analyser l'utilisation de la citation

### Subtask:
Répondez aux questions suivantes : Dans quel type de communication interne citer serait-il plus approprié que paraphraser ? Quand est-ce que citer peut présenter un risque ?


```markdown
### Quand citer est plus approprié que paraphraser en communication interne ?

Citer directement est plus approprié que paraphraser dans les types de communication interne suivants, où la précision, l'autorité ou l'authenticité du message original est primordiale :

*   **Documents juridiques et conformité :** Lors de la communication de lois, de règlements, de clauses contractuelles ou de politiques d'entreprise strictes. Citer garantit que les termes exacts et les nuances juridiques sont préservés, évitant toute interprétation erronée qui pourrait avoir des conséquences légales ou de conformité.

    *Exemple :* L'annonce d'une nouvelle politique de protection des données exigeant une citation directe des extraits clés pour s'assurer que les employés comprennent leurs obligations exactes.

*   **Politiques et procédures officielles :** Lorsque l'adhésion exacte à une politique ou une procédure est essentielle pour la sécurité, l'efficacité opérationnelle ou la qualité. Citer peut éviter la dilution ou la modification accidentelle du sens original.

    *Exemple :* Les instructions détaillées pour l'utilisation d'un nouvel équipement de sécurité ou les protocoles d'urgence, où chaque mot compte.

*   **Communications de la haute direction sur des sujets critiques :** Les messages originaux des dirigeants sur des décisions stratégiques, des changements majeurs ou des valeurs fondamentales peuvent perdre de leur impact et de leur autorité s'ils sont paraphrasés. Citer préserve la voix et l'intention du leadership.

    *Exemple :* Une déclaration du PDG concernant une fusion ou une acquisition majeure, où les mots exacts peuvent refléter la vision de l'entreprise.

*   **Témoignages ou retours clients authentiques :** Pour valoriser la crédibilité et l'émotion de retours d'expérience, qu'ils soient positifs ou négatifs. La citation directe renforce l'authenticité.

    *Exemple :* Un témoignage d'un client satisfait utilisé dans une communication interne pour motiver les équipes.

*   **Rapports d'incident ou analyses techniques :** Lorsque les détails techniques ou les observations spécifiques sont cruciaux pour l'analyse, la compréhension ou la résolution de problèmes. Paraphraser pourrait introduire des inexactitudes.

    *Exemple :* Un extrait d'un rapport de cyber-sécurité décrivant une vulnérabilité spécifique ou une erreur de code.

Dans ces contextes, la valeur de la précision l'emporte sur les avantages potentiels d'une paraphrase simplifiée, car une mauvaise interprétation pourrait avoir des conséquences graves.

```markdown
### Quand citer peut présenter un risque en communication interne ?

Bien que la citation directe soit essentielle dans certains cas, elle présente aussi des risques en communication interne si elle est utilisée de manière inappropriée ou excessive :

*   **Perte de clarté et de compréhension :** Le langage utilisé dans une citation originale peut être trop technique, jargonnesque, formel ou complexe pour le public visé. Citer sans explication ou simplification peut entraîner une mauvaise compréhension ou une confusion.

    *Exemple :* Citer des extraits juridiques complexes à des employés non juristes sans en simplifier le sens.

*   **Manque de pertinence et surcharge d'information :** Une citation peut contenir des informations superflues ou des détails qui ne sont pas directement pertinents pour le message principal que l'on souhaite transmettre. Cela peut diluer l'impact du message et fatiguer le lecteur.

    *Exemple :* Citer un long paragraphe d'un rapport technique alors que seule une phrase est cruciale pour la décision.

*   **Manque d'adaptation au ton et au contexte :** Une citation peut avoir un ton ou un style qui ne correspond pas à la communication interne générale ou au public spécifique. Un ton trop formel dans une communication informelle peut paraître rigide et créer une distance.

    *Exemple :* Utiliser une citation très formelle d'un document de politique dans un e-mail d'équipe informel.

*   **Détournement de sens ou interprétation erronée :** Une citation sortie de son contexte peut être mal interprétée ou son sens peut être déformé, même involontairement. Le lecteur pourrait se concentrer sur des éléments secondaires ou mal comprendre l'intention initiale.

    *Exemple :* Citer une phrase d'une déclaration de leadership sans le contexte complet des discussions ou des justifications, conduisant à des rumeurs ou des malentendus.

*   **Manque d'engagement :** Les citations directes, surtout si elles sont nombreuses ou longues, peuvent donner l'impression que l'émetteur du message ne s'est pas approprié l'information ou n'a pas fait l'effort de la synthétiser. Cela peut réduire l'engagement du public.

*   **Problèmes de confidentialité ou de sensibilité :** Certaines citations peuvent contenir des informations sensibles, confidentielles ou des formulations qui, si elles sont répétées textuellement, pourraient causer des problèmes (ex. fuites d'informations, exacerbation de tensions).

    *Exemple :* Citer directement une critique négative virulente d'un employé dans une communication plus large, risquant de créer du ressentiment.

Dans ces situations, une paraphrase bien pensée permet de reformuler l'information de manière plus claire, plus concise, mieux adaptée au public et au contexte, tout en conservant l'essentiel du message original.
